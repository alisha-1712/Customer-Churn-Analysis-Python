```python
# 03_seaborn_eda.ipynb


import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")

df = pd.read_csv("../data/customer_churn_cleaned.csv")
```

```python
# churn distribution
plt.figure(figsize=(6,4))
sns.countplot(x=df['churn_label'])
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Churn Rate by Tenure Group

df['tenure_group'] = np.where(
    df['tenure_months'] < 6, 'New',
    np.where(df['tenure_months'] <= 24, 'Mid', 'Loyal')
)

df['tenure_group'].value_counts(normalize=True)

plt.figure(figsize=(7,4))
sns.barplot(
    x=df['tenure_group'],
    y=df['churn'],
    estimator='mean',
    ci=None
)
plt.title("Churn Rate by Tenure Group")
plt.xlabel("Tenure Group")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# Churn Rate by Contract Type
plt.figure(figsize=(8,4))
sns.barplot(
    x=df['contract_type'],
    y=df['churn'],
    estimator='mean',
    ci=95
)
plt.title("Churn Rate by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Churn Rate")
plt.xticks(rotation=30)
plt.show()
```

```python
# Monthly Logins vs Churn 
plt.figure(figsize=(7,4))
sns.boxplot(
    x=df['churn_label'],
    y=df['monthly_logins']
)
plt.title("Monthly Logins by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Monthly Logins")
plt.show()
```

```python
# Days Since Last Login vs Churn (Boxplot)
plt.figure(figsize=(7,4))
sns.boxplot(
    x=df['churn_label'],
    y=df['last_login_days_ago']
)
plt.title("Days Since Last Login by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Days Since Last Login")
plt.show()
```

```python
# Key Observations
# Churn is heavily concentrated among new customers.
# Engagement frequency (monthly logins) shows meaningful separation.
# Contract type does not strongly influence churn.
# No single metric explains churn; combined signals are more effective.
```
