```python

# 01_setup_and_cleaning.ipynb

import pandas as pd
import numpy as np

# Display settings
pd.set_option('display.max_columns', None)

# Load dataset
df = pd.read_csv("../data/customer_churn.csv")

# Quick look
df.head()
```

```python
# Basic structure
df.shape
df.info()
# Missing values check
df.isna().sum()
```

```python
# Create churn label for readability
df['churn_label'] = df['churn'].map({0: 'No', 1: 'Yes'})

df['churn_label'].value_counts()
```

```python
# Save cleaned base dataset 
df.to_csv("../data/customer_churn_cleaned.csv", index=False)
```
